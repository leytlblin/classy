```python
import json
```

```python
tokeni={}
chehson=[]
for token in cheh:
  tokeni={"lemma":token.lemma_, "word":token.text, "pos":token.pos_}
  if token.pos_ != 'SPACE' and token.pos_ != 'PUNCT' and token.lemma_ not in ['?', '!', '"', '(', ')', '-']:
    chehson.append(tokeni)
print(chehson)

```

```python
for token in chehson:
  if token['pos'] != 'SPACE' and token['pos'] != 'PUNCT' and token['lemma'] not in ['?', '!', '"', '(', ')', '-']:
    print(token['lemma'])

```

```python
with open('c.jsonl', 'w', encoding = 'utf-8') as f:
    for d in chehson:
        json.dump(d, f, ensure_ascii=False)
        f.write('\n')
```
